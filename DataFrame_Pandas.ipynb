#Chuyển dạng dữ liệu float sang String để xử lý

```python
data3['lat-lon']=data3['lat-lon'].astype(str)
```

#Thêm cột + Tách Split 1 đoạn dữ liệu từ cột cũ

```python
data3['lat'] = data3['lat-lon'].apply(lambda i: i.split(',')[0])
data3
```

```python
data3 = data3.drop('lat',axis=1)
data3
```

#Tạo bảng mới từ Bảng có sẵn

```python
data4 = pd.DataFrame(data3)
data4['lat'] = data4['lat-lon'].str.split(",",expand= True)[0]
data4['lon'] = data4['lat-lon'].str.split(",",expand= True)[1]
data4


```
