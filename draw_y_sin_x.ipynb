```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.title('y= sin(x)\n')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
